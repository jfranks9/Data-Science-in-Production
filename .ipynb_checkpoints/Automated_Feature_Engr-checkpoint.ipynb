```python
from sklearn.model_selection import train_test_split
import pandas as pd
import featuretools as ft
from featuretools import Feature

#dataset
game_df = pd.read_csv('../nhl-game-data/game.csv')  


#Pulling first 50000 rows due to memory issues in EC2 instance

plays_df = pd.read_csv('../nhl-game-data/game_plays.csv', error_bad_lines=False, nrows=50000)

plays_df = plays_df.drop(['secondaryType', 'periodType', 'dateTime', 'rink_side'],axis=1).fillna(0)

#plays_df.head()
game_df.head()
```

```python
#Building the model
es = ft.EntitySet(id="plays")
es = es.entity_from_dataframe(entity_id="plays", dataframe=plays_df,index="play_id",
    variable_types = { "event": ft.variable_types.Categorical, "description": ft.variable_types.Categorical})

f1 = Feature(es["plays"]["event"])
f2 = Feature(es["plays"]["description"])

encoded, defs = ft.encode_features(plays_df,[f1,f2], top_n=10)
encoded.reset_index(inplace=True)
encoded.head()

es = ft.EntitySet(id="plays")  
es = es.entity_from_dataframe(entity_id="plays",  dataframe=encoded, index="play_id")  
es = es.normalize_entity(base_entity_id="plays",  new_entity_id="games", index="game_id")  

features,transform=ft.dfs(entityset=es,  target_entity="games",max_depth=2)  
features.reset_index(inplace=True)  
features.head()

```

```python
import framequery as fq
import numpy as np

features = fq.execute("""  
    
     SELECT f.*  
        ,case when g.type = 'P' then 1  else 0 end as label
        
    FROM features f
   
    JOIN game_df g
       on f.game_id = g.game_id  
""") 
print(features)
```

```python
#Issue with all games being post-season or 'P' not sure why
from sklearn.linear_model import LogisticRegression  
from sklearn.metrics import roc_auc_score  

#create inputs for sklearn  
y = features['label'] 

X = features.drop(['label', 'game_id'], axis=1).fillna(0)  

#train a classifier  
lr = LogisticRegression()  
model = lr.fit(X, y)  

#Results  
#print("Accuracy: " + str(model.score(X, y)))  
#print("ROC" + str(roc_auc_score(y,model.predict_proba(X)[:,1]))) 

#print(y)
```
